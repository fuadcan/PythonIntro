```python
import requests
import re
from time import sleep
from random import uniform
```

```python
url   = "http://www.istanbulfirmarehber.com/rehber/ustid/68/altid/0/ilceid/5.htm"
sleep(uniform(2,15))
resps = requests.get(url)
html  = resps.text
print(html)
```

```python
telefonlar = re.findall('Tel.*?\([0-9]+\)[0-9]+',html)
for telefon in telefonlar:
    print(telefon)
```

### BeautifulSoup4

```python
import bs4
soup = bs4.BeautifulSoup(html,'html.parser')
```

```python
for link in soup.find_all('a'):
    print(link)
```

```python
for link in soup.find_all('td'):
    print(link.text)
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    # print(link.text)
    # print(re.sub('\s+',' ',re.sub('\n','|',link.text)))
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if not re.sub('\s+','',part) == '':
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if re.search('\S',part):
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    print('|'.join(parts))        
```

### txt'ye yazma 

```python
firmalar = open('outputs/firmalarscraped.txt','w')
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmalar.write('|'.join(parts) + '\n')
firmalar.close()
```
