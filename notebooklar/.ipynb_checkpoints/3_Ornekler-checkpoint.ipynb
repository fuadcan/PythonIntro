## Pandas.apply()

```python
bugun = datetime.datetime.today().date()
IK['Age'] = IK['DOB'].apply(lambda x: (bugun - pd.Timestamp(x).date()).days/365)
```

```python
IK['YearWorked'] = IK['Date of Hire'].apply(lambda x: (bugun - pd.Timestamp(x).date()).days/365)
```

```python
sum(IK['Sex'].apply(lambda x: x.lower()) == 'female')
```
