### Günlük toplam ortalama Satışlar ve Adetler

```python
gunluk_satis = satislar.groupby("Ship Date")[.toplanacak değerler.].agg({.uygulanacak fonksiyonlar.}).reset_index()
```

```python
# verinin satır index'ini tarihlerle değiştir
gunluk_satis.index = gunluk_satis['Ship Date']
```

```python
# ilk kolonu (Ship Date) sil
gunluk_satis.iloc[:,1:]
```

### Data yazma

#### CSV'ye yazma

```python
gunluk_satis.to_csv('outputs/gunluksatis.csv',index=False,sep=',',header=True)
```

#### Excel'e yazma

```python
gunluk_satis.to_excel('outputs/gunluksatis.xlsx',index=False,header=True)
```

#### JSON'a yazma

```python
gunluk_satis.to_excel('outputs/gunluksatis.xlsx',index=False)
```

#### SQLite'a yazma

```python
import sqlite3
# from sqlalchemy import create_engine
# engine = create_engine('postgresql://'+user+':'+password+'@'+host+'/'+dbname)
engine = sqlite3.connect("outputs/satislar.db")
gunluk_satis.to_sql(name='satislar', con=engine)
```
