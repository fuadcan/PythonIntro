### Dict to CSV

Bazı durumlarda dictionary objesini yazmak daha iyi sonuçlar verebilir. Örneğin verinin başlık satırının doğru yazıldığından emin olunmak isteniyorsa başlıklar ile verinin denk getirilmesini `dictwriter` sağlayabilir.

```python
maillist[1:]
```

```python
data_dict = []
basliklar = maillist[0]
data      = maillist[1:]
```

```python
len(basliklar)
```

```python
data[0]
```

```python
for line in data:
    temp = {basliklar[i]: line[i] for i in range(len(basliklar))}
    data_dict.append(temp)
data_dict
```

### Alternatif (tek satır)

```python
data_dict = [{basliklar[i]: line[i] for i in range(len(basliklar))} for line in data]
```

```python
data_dict[1]
```

```python
# dictionary yazma
import csv
with codecs.open('outputs/mailler.csv', 'w', encoding='utf-8') as mailler:
    writer = csv.DictWriter(mailler,fieldnames=basliklar)
    # başlık satırını yaz (ipucu: writeheader())
    writer.writeheader()
    for dct in data_dict:
        # her bir satırı yaz (ipucu: write.row)
        writer.writerow(dct)
```

#### CSV okuma

```python
# dictionary yazma
import csv
with codecs.open('outputs/mailler.csv', 'r', encoding='utf-8') as mailler:
    reader = csv.DictReader(mailler)
    for row in reader:
        print(row['Ad'], row['email'])
```

### JSON'a yazma

```python
data_dict
```

```python
import json, os

with open('outputs/maillist.json', 'w') as fh:
    fh.write('[')
    for line in data_dict:
        # dict'i doldur
        # Yaz
        # ipucu: her bir iterasyonda bir alt satıra geç
        ## dictionary'yi yaz ve yanına bir virgül koy.
        ## Bir sonraki iterasyonda yeniden bir alt
        ## satıra geçecek
        # ipucu: json.dump()
        fh.write('\n')
        json.dump(line,fh)
        fh.write(',')
    # Son iterasyonda eklenen gereksiz virgül'ü sil
    fh.seek(fh.tell()-1)
    fh.truncate()
    # Bir satır aşağıya
    fh.write('\n]')
```

Yukarıdaki örnekte `cur.execute` her bir iterasyonda yazılacak değeri hazırlıyor ve `conn.commit()` anlık hafızada bulunan bu veriyi _firmalar.db_'ye işliyor. Ancak bu işlemi her iterasyonda yapmak yazma işlemini yavaşlatır. Bunun yerine 10 (ya da 1000) iterasyonda bir `conn.commit()` emrini verirsek hafızada 10 (ya da 1000) işlem biriktirdikten sonra yazma işlemini gerçekleştirebiliriz. Ancak her iterasyonda işlenen veriler büyüdükçe biriken veriler de büyüyecektir ve bilgisayarınızın RAM'ini dolduracak kadar birikirse hata mesajı verecektir.
