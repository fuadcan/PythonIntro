### BeautifulSoup4

```python
import bs4
soup = bs4.BeautifulSoup(html,'html.parser')
```

```python
for link in soup.find_all('a'):
    print(link)
    print('------')
```

```python
soup.prettify('utf-8')
```

```python
for link in soup.find_all('td'):
    print(link.text)
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    # print(link.text.strip())
    print(re.sub('\s+',' ',re.sub('\n','|',link.text)))
    # print(re.sub('\s+',' ',re.sub('\n','|',link.text)))
    print('--------')
```

```python
for adacik in soup.find_all('td', attrs={'class':'ch'}):
    # her bir td'yi \n'lerden ayır
    parts = adacik.text.split('\n')
    # print(parts)
    # her parça için eğer içinde kelime karakteri varsa listeye ekle, yoksa atla
    for part in parts:
        if re.search('\S',part):
            print(part.strip())
    print('----')
```

```python
from time import sleep
import codecs
firmalar = open('outputs/firmalarscraped.txt','w', encoding='utf-8')
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmalar.write('|'.join(parts) + '\n')
firmalar.close()
```
