```python
import requests
```

```python
url = "http://www.istanbulfirmarehber.com/rehber/ustid/35/altid/94.htm"
sonuc = requests.get(url)
sonuc
```

```python
sonuc.text
```

```python
import re

telefonlar = re.findall('\([0-9]+\)[0-9]+',sonuc.text)
for telefon in telefonlar:
    telefon = telefon.replace("(","").replace(")","")
    print(telefon)
```

```python
import re

telefonlar = re.findall('\([0-9]+\)[0-9]+',sonuc.text)
for telefon in telefonlar:
    telefon = re.sub('\(|\)','',telefon)
    print(telefon)
```

```python
from bs4 import BeautifulSoup
html_kodu = sonuc.text

#çorbayı ateşe ver:
soup = BeautifulSoup(html_kodu)

for link in soup.find_all('td', attrs={'class':'ch'}):
    print(link,'\n')
```

```python
from bs4 import BeautifulSoup
html_kodu = sonuc.text

#çorbayı ateşe ver:
soup = BeautifulSoup(html_kodu)

for link in soup.find_all('td', attrs={'class':'ch'}):
    firma_bilgileri = link.text
    firma_bilgileri = firma_bilgileri.split('\n')
    for element in firma_bilgileri:
        if re.search('\S',element):
            print(element.strip())
    print('\n')
        
```

```python
from bs4 import BeautifulSoup
html_kodu = sonuc.text

#çorbayı ateşe ver:
soup = BeautifulSoup(html_kodu)

for link in soup.find_all('td', attrs={'class':'ch'}):
    firma_bilgileri = link.text
    firma_bilgileri = firma_bilgileri.split('\n')
    icerik = []
    for element in firma_bilgileri:
        if re.search('\S',element):
            icerik.append(element.strip())
    print('|'.join(icerik))
```

```python
from bs4 import BeautifulSoup
html_kodu = sonuc.text

out = open('firma_bilgileri.txt','w')
#çorbayı ateşe ver:
soup = BeautifulSoup(html_kodu)

for link in soup.find_all('td', attrs={'class':'ch'}):
    firma_bilgileri = link.text
    firma_bilgileri = firma_bilgileri.split('\n')
    icerik = []
    for element in firma_bilgileri:
        if re.search('\S',element):
            icerik.append(element.strip())
    out.write('|'.join(icerik) +  '\n')
out.close()
```
