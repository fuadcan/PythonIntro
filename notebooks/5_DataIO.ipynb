```python
fh = open('firmalar_processed.txt')
outfile = open('leventler.txt','w')
counter = 0
for line in fh:
    if counter <= leventler:
        outfile.write(line)
    counter += 1
fh.close()
outfile.close()
```

### TelefonNumaralari.txt

```python
import re
regex_kodu = "Tel\s:\s\([0-9]+\)[0-9]+"
fh = open('firmalar_processed.txt')
for line in fh:
    line = line.rstrip()
    if line.startswith('Tel'):
        print(re.findall(regex_kodu,line))
fh.close()
```

```python
import re
fh = codecs.open('firmalar_processed.txt',encoding = 'utf-8')
for line in fh:
    line = line.rstrip()
    if line.startswith('Tel'):
        print(re.findall('Tel\s:\s\([0-9]+\)[0-9]+',line))
fh.close()
```

```python
import re
fh = open('firmalar_processed.txt')
for line in fh:
    line = line.rstrip()
    if line.startswith('Tel'):
        print(re.findall('Tel.*\([0-9]+\)[0-9]+',line))
fh.close()
```
