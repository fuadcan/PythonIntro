```python
import sqlite3,json,os
```

```python
def db_creator(dbpath, dbname):
        conn = sqlite3.connect(dbpath + '/' + dbname)
        cur  = conn.cursor()
        cur.execute('''CREATE TABLE lyrics
                     (name TEXT, lyrics TEXT, artist TEXT, genre TEXT, album TEXT,year INTEGER, source TEXT)''')
        cur.execute('''CREATE UNIQUE INDEX id ON lyrics (name, artist, album)''')

        conn.commit()    
        conn.close()
   
```

```python
db_creator('output','lyrics.db')
```

```python
def album_writer(song_info_list,dbdest):
    conn = sqlite3.connect(dbdest)
    cur  = conn.cursor()

    for s in song_info_list:
        # Yazma
        cur.execute("INSERT OR IGNORE INTO lyrics VALUES (?,?,?,?,?,?,?)", 
                    (s['name'],s['lyrics'],s['artist'],s['genre'],s['album'],s['year'],s['source']))
        print('[LOG] Written songs for %s - %s' % (s['album'], s['artist']))
    conn.commit()    
    conn.close()
```

### Text Cleaner

```python
def text_cleaner(text):
    substitutes = [("'ll "," will "),("'m "," am "),("'d "," would "),("'re "," are "),
               ("'s "," is "),("'t "," not "),("'ve "," have "), ("'cause ", " because ")]
    substitutes = substitutes + [("`ll "," will "),("`m "," am "),("`d "," would "),("`re "," are "),
               ("`s "," is "),("`t "," not "),("`ve "," have "), ("`cause ", " because ")]
    for sub in substitutes:
        text = text.replace(sub[0],sub[1])
    text = re.sub('\W',' ',text)
    text = text.lower()

    return(text)
```

```python
for song in prog:
    print(text_cleaner(song[2]))
```

```python
from collections import defaultdict
progfd = defaultdict(int)
for song in prog:
    song  = text_cleaner(song[2])
    words = song.split()
    for word in words:
        progfd[word] += 1
```

### Stopwords

```python
import requests
resp = requests.get("https://gist.githubusercontent.com/sebleier/554280/raw/7e0e4a1ce04c2bb7bd41089c9821dbcf6d0c786c/NLTK's%2520list%2520of%2520english%2520stopwords")
content = resp.content
stopwords = content.decode('utf-8').split('\n')
```

```python
for word in stopwords:
    if word in progfd:
        del(progfd[word])
```
