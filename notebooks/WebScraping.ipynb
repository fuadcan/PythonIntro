```python
from bs4 import BeautifulSoup as bs
import requests, re

url = "http://www.istanbulfirmarehber.com/rehber/ustid/68/altid/0/ilceid/5.htm"
html = requests.get(url)
soup = bs(html.text)
```

```python
for link in soup.find_all('a'):
    print(link)
```

```python
for link in soup.find_all('td'):
    print(link.text)
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    # print(link.text)
    print(re.sub('\s+',' ',re.sub('\n','|',link.text)))
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if not re.sub('\s+','',part) == '':
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if re.search('\S',part):
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    print('|'.join(parts))        
```

```python
firmalar = open('firmalarscraped.txt','w')
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmalar.write('|'.join(parts) + '\n')
firmalar.close()
```
