```python
from bs4 import BeautifulSoup as bs
import bs4
import requests, re

url = "http://www.darklyrics.com/lyrics/haken/aquarius.html"
html = requests.get(url)
soup = bs(html.content,'html.parser')
```

```python
print(soup.prettify)
```

```python
for link in soup.find_all('td'):
    print(link.text)
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    # print(link.text)
    print(re.sub('\s+',' ',re.sub('\n','|',link.text)))
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if not re.sub('\s+','',part) == '':
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    for part in parts:
        if re.search('\S',part):
            print(part.strip())
    print('\n')
```

```python
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    print('|'.join(parts))        
```

### txt ve csv'ye yazma 

```python
firmalar = open('outputs/firmalarscraped.txt','w')
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmalar.write('|'.join(parts) + '\n')
firmalar.close()
```

```python
with open('outputs/firmalarscraped.txt','w') as firmalar:
    for link in soup.find_all('td', attrs={'class':'ch'}):
        parts = link.text.split('\n')
        parts = [part.strip() for part in parts if re.search('\S',part)]
        firmalar.write('|'.join(parts) + '\n')
```

```python
import csv
with open('outputs/firmalarscraped.csv', 'w', newline='') as firmalar:
    firmawriter = csv.writer(firmalar, delimiter=';',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for link in soup.find_all('td', attrs={'class':'ch'}):
        parts = link.text.split('\n')
        parts = [part.strip() for part in parts if re.search('\S',part)]
        print(parts)
        firmawriter.writerow(parts)
```

```python
import csv    

with open('outputs/firmalarscraped.csv', 'w', newline='') as firmalar:
    fields = ['Firma','Adres','Tel','Web']
    firmawriter = csv.DictWriter(firmalar, fieldnames=fields)
    firmawriter.writeheader()
    for link in soup.find_all('td', attrs={'class':'ch'}):
        parts = link.text.split('\n')
        parts = [part.strip() for part in parts if re.search('\S',part)]
        firmdict = dict()
        # Dictionary oluşturma
        for i in range(len(parts)):
            firmdict[fields[i]] = parts[i]
        # Yazma
        firmawriter.writerow(firmdict)
```

### JSON'a yazma

```python
import json, os

with open('outputs/firmalar.json', 'w') as fh:
    fh.write('[')
    for link in soup.find_all('td', attrs={'class':'ch'}):
        parts = link.text.split('\n')
        parts = [part.strip() for part in parts if re.search('\S',part)]
        firmdict = dict()
        # Dictionary oluşturma
        for i in range(len(parts)):
            firmdict[fields[i]] = parts[i]
        # Yazma
        fh.write('\n')
        json.dump(firmdict, fh)
        fh.write(',')
    # Son iterasyonda eklenen gereksiz virgül'ü sil
    fh.seek(fh.tell()-1)
    fh.truncate()
    # Bir satır aşağıya
    fh.write('\n]')
```

### SQL'e yazma 

```python
import sqlite3
conn = sqlite3.connect('outputs/firmalar.db')

cur  = conn.cursor()

cur.execute('''CREATE TABLE firmalar
             (Firma text, Adres text, Tel text, Web text)''')

for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmdict = dict()
    # Dictionary oluşturma
    parts = parts + [None] * (4 - len(parts))
    # Yazma
    cur.execute("INSERT INTO firmalar VALUES (?,?,?,?)", parts)
    conn.commit()
conn.close()
```

Yukarıdaki örnekte `cur.execute` her bir iterasyonda yazılacak değeri hazırlıyor ve `conn.commit()` anlık hafızada bulunan bu veriyi _firmalar.db_'ye işliyor. Ancak bu işlemi her iterasyonda yapmak yazma işlemini yavaşlatır. Bunun yerine 10 (ya da 1000) iterasyonda bir `conn.commit()` emrini verirsek hafızada 10 (ya da 1000) işlem biriktirdikten sonra yazma işlemini gerçekleştirebiliriz. Ancak her iterasyonda işlenen veriler büyüdükçe biriken veriler de büyüyecektir ve bilgisayarınızın RAM'ini dolduracak kadar birikirse hata mesajı verecektir. Bizim örneğimizde verinin tamamı bile RAM'i doldurmasa da işlemi 10 iterasyonda 1 yapacak şekilde uygulayacağız:

```python
import sqlite3
conn = sqlite3.connect('outputs/firmalar.db')

conn.execute('DROP TABLE firmalar') # daha önce oluşturduğumuz tabloyu sil
cur  = conn.cursor()

cur.execute('''CREATE TABLE firmalar
             (Firma text, Adres text, Tel text, Web text)''')
counter = 1
for link in soup.find_all('td', attrs={'class':'ch'}):
    parts = link.text.split('\n')
    parts = [part.strip() for part in parts if re.search('\S',part)]
    firmdict = dict()
    # Dictionary oluşturma
    parts = parts + [None] * (4 - len(parts))
    # Yazma
    cur.execute("INSERT INTO firmalar VALUES (?,?,?,?)", parts)
    if counter % 10 == 0: 
        conn.commit()
        print('commited at', counter)
    counter += 1
conn.close()
```
