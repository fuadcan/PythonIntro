## Yaş hesapla

```python
import datetime
dogumTarihleri = IK['DOB'].values
dogumTarihleri = [pd.Timestamp(dogum).date() for dogum in dogumTarihleri]
```

```python
type(datetime.datetime.now())
```

```python
def yas_hesapla(dogumTarihi):
    # bugünün tarihi datetime.date objesi olarak
    bugun = datetime.datetime.now()
    # tarihler arası geçen gün sayısının 365'e oranı ipucu: (dn - d0).days
    yas   = (bugun.date() - dogumTarihi).days/365
    return(yas)
```

```python
# tüm doğum tarihlerine uygula
yas_listesi = []
for dogumTarihi in dogumTarihleri:
    yas = yas_hesapla(dogumTarihi)
    yas_listesi.append(yas)
```

## Alternatif

```python
yas_listesi = [yas_hesapla(dogumTarihi) for dogumTarihi in dogumTarihleri]
```

## Pandas.apply()

```python
bugun = datetime.datetime.today().date()
IK['Age'] = IK['DOB'].apply(lambda x: (bugun - pd.Timestamp(x).date()).days/365)
```

```python
IK['YearWorked'] = IK['Date of Hire'].apply(lambda x: (bugun - pd.Timestamp(x).date()).days/365)
```
