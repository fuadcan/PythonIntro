# Neden Python

## 1. Python Etkindir (efficient)

 - Python'un mimarisi hız açısından diğer dillerin gerisine düşmesinin önüne geçmektedir
 - Hızlıdır ve az hafıza kaplar
 - Fonksiyonel programlama dilidir, her hesaplanan veri hafızada haritalama yolu ile tutulur (swallow copy)

## 2. Community

 - Python'un çok geniş bir komünitesi vardır
 - Yeni çıkan modeller, algoritmalar hızlıca adapte edilir
 - Problem çıktığında hızla çözüm bulunabilir
 - Çok fazla paket desteği vardır
 
## 3. Open Source

 - Python açık kaynak kodludur.
 - Bu yüzden hem hızlıca adapte olur hem de ürünlerin çok büyük çoğunluğu açık kaynaklıdır ve ücretsizdir.

## 4. Easy Syntax / Steep Learning Curve

 - Sentaksı çok kolaydır
 - Diğer dillerle çok satırda yazılan bir kod Python'da çok az satırda yazılabilir
 - Bu yüzden öğrenmesi kolaydır ve öğrenme grafiği hızlı yükselir

## 5. Popülarite ve Maaş
 - Python'un popülaritesi gittikçe artmaktadır
 - ML kütüphaneleri ve kullanım kolaylığı sebebiyle son yıllarda developerlar tarafından daha sık tercih edilmektedir
